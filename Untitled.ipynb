```python
data['major'] = data.major.str.lower()
data['major'] = data.major.str.strip()
data['major'] = data.major.str.replace('\W', ' ', regex=True)
data['major'] = data.major.str.replace('  +', ' ', regex=True)
data['major'] = data.major.str.split()
data['major'] = data.major.str[0]
```

```python
from spellchecker import SpellChecker
spell = SpellChecker(distance=1)
def Correct(x):
    return spell.correction(str(x))
data['major'] = data['major'].apply(Correct)
```

```python
depression_answers = ['Q3A', 'Q5A', 'Q10A', 'Q13A', 'Q16A', 'Q17A', 'Q21A', 'Q24A', 'Q26A', 'Q31A', 'Q34A', 'Q37A', 'Q38A', 'Q42A']
anxiety_answers = ['Q2A', 'Q4A', 'Q7A', 'Q9A', 'Q15A', 'Q19A', 'Q20A', 'Q23A', 'Q25A', 'Q28A','Q30A', 'Q36A', 'Q40A', 'Q41A']
stress_answers = ['Q1A', 'Q6A', 'Q8A', 'Q11A', 'Q12A', 'Q14A', 'Q18A', 'Q22A', 'Q27A', 'Q29A', 'Q32A', 'Q33A', 'Q35A', 'Q39A']

depression_time = [i.strip('A') + 'E' for i in depression_answers]
anxiety_time = [i.strip('A') + 'E' for i in anxiety_answers]
stress_time = [i.strip('A') + 'E' for i in stress_answers]
```

```python
data['dep_score'] = data[depression_answers].sum(axis=1)
data['anx_score'] = data[anxiety_answers].sum(axis=1)
data['str_score'] = data[stress_answers].sum(axis=1)

data['dep_time'] = data[depression_time].sum(axis=1)
data['anx_time'] = data[anxiety_time].sum(axis=1)
data['str_time'] = data[stress_time].sum(axis=1)

data['total_time(ms)'] = data['dep_time'] + data['anx_time'] + data['str_time']

data['dep_time'] = round(100 * data['dep_time']/data['total_time(ms)'], 2)
data['anx_time'] = round(100 * data['anx_time']/data['total_time(ms)'], 2)
data['str_time'] = round(100 * data['str_time']/data['total_time(ms)'], 2)

all_questions = depression_answers + anxiety_answers + stress_answers
question_position = [i.strip('A') + 'I' for i in all_questions]

data = data.drop(columns=depression_answers + anxiety_answers + stress_answers 
                         + depression_time + anxiety_time + stress_time + question_position
                         + ['total_time(ms)', 'introelapse', 'testelapse', 'surveyelapse'], errors = 'ignore')
```

```python
data['dep_cat']=np.select([data['dep_score'] < 10, 
                           data['dep_score'].between(10,14),
                           data['dep_score'].between(14,21),
                           data['dep_score'].between(21,28),
                           data['dep_score'] >= 28],
                           [0, 1, 2, 3, 4])

data['anx_cat'] = np.select([data['anx_score'] < 8, 
                             data['anx_score'].between(8,10),
                             data['anx_score'].between(10,15),
                             data['anx_score'].between(15,20),
                             data['anx_score'] >= 20],
                             [0, 1, 2, 3, 4])

data['str_cat'] = np.select([data['str_score'] < 15, 
                             data['str_score'].between(15,19),
                             data['str_score'].between(19,26),
                             data['str_score'].between(26,34),
                             data['str_score'] >= 34],
                             [0, 1, 2, 3, 4])
```

```python
normal_tipis = ['TIPI' + str(i) for i in range(1,11,2)]
reverse_tipis = ['TIPI' + str(i) for i in range(1,11) if i % 2 == 0]
```

```python
edu_dict = {1: 'Below High School',
            2: 'High School',
            3: 'University Degree',
            4: 'Graduate Degree'}

urb_dict = {1: 'Rural',
            2: 'Suburban',
            3: 'Urban'}

race_dict = {10:'Asian', 
             20:'Arab', 
             30:'Black', 
             40:'Indigenous Australian', 
             50:'Native American', 
             60:'White', 
             70:'Other'}

hand_dict = {1: 'Right',
             2: 'Left',
             3: 'Both'}

relig_dict = {1:'Agnostic', 
              2:'Atheist', 
              3:'Buddhist', 
              4:'Christian (Catholic)', 
              5:'Christian (Mormon)', 
              6:'Christian (Protestant)', 
              7:'Christian (Other)', 
              8:'Hindu', 
              9:'Jewish', 
              10:'Muslim', 
              11:'Sikh', 
              12:'Other'}

orient_dict = {0:'Other',
               1:'Heterosexual', 
               2:'Bisexual', 
               3:'Homosexual', 
               4:'Asexual', 
               5:'Other'}

gend_dict = {1: 'M',
             2: 'F',
             3: 'O'}

marr_dict = {1:'Never',
             2:'Currently',
             3:'Previously'}

data['education'] = data.education.replace(edu_dict)
data['urban'] = data.urban.replace(urb_dict)
data['race'] = data.race.replace(race_dict)
data['hand'] = data.hand.replace(hand_dict)
data['religion'] = data.religion.replace(relig_dict)
data['orientation'] = data.orientation.replace(orient_dict)
data['gender'] = data.gender.replace(gend_dict)
data['married'] = data.married.replace(marr_dict)
```
