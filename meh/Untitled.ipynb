```python
games = games.set_index('ResponseID')
```

```python
games = games.drop(columns=['QueryID', 'LinuxRecReqsText', 'MacRecReqsText', 'PCRecReqsText', 'Website', 
                            'ShortDescrip', 'DetailedDescrip', 'HeaderImage', 'LegalNotice', 'QueryName',
                            'SupportEmail', 'SupportURL', 'AboutText', 'DRMNotice', 'ExtUserAcctNotice',
                            'Reviews', 'Background', 'PCMinReqsText', 'LinuxMinReqsText', 'MacMinReqsText',
                            'PriceCurrency', ' ', 'PublisherCount', 'SteamSpyPlayersVariance', 
                            'SteamSpyOwnersVariance'], errors='ignore')
games['ReleaseDate'] = pd.to_datetime(games['ReleaseDate'], format='%b %d %Y', errors='coerce')
games = games.dropna(subset=['ReleaseDate'])
```

```python
games['SupportedLanguages'] = games['SupportedLanguages'].str.strip()
remove = ['\*', '\[', '\]', '\/', '[b]', 'languages', 'with', 'full','audio','support', 'all', '\)', '\(', 'text', 'only']
games['SupportedLanguages'] = games['SupportedLanguages'].replace('|'.join(remove),'',regex=True)
games['SupportedLanguages'] = games['SupportedLanguages'].replace({'RussianEnglishSpanishFrenchJapaneseCzech':'Russian English Spanish French Japanese Czech',
                                                                   'EnglishRussianSpanishJapaneseCzech':'English Russian Spanish Japanese Czech',
                                                                   'HungarianPolish':'Hungarian Polish', 
                                                                   'Araic':'Arabic', 
                                                                   '  +': ' ',
                                                                   'Simplified Chinese':'SimplifiedChinese', 
                                                                   'Traditional Chinese':'TraditionalChinese'
                                                                   },regex=True) 
```

```python
games['SupportedLanguages'] = games['SupportedLanguages'].apply(lambda x: x.split(' '))
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer(sparse_output=True)

games = games.join(
                pd.DataFrame.sparse.from_spmatrix(
                mlb.fit_transform(games.pop('SupportedLanguages')),
                index=games.index,
                columns=mlb.classes_))
```
