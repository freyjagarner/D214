### Getting DASS Scores

First we will get our DASS scores using the scoring guide found [here](https://neurocogsystem.com/wp-content/uploads/2021/02/DASS-42-Scoring.pdf).

According to this scoring guide which is based off of the DASS Manual questions number 3, 5, 10, 13, 16, 17, 21, 24, 26, 31, 34, 38 and 42 are the depression based questions. The anxiety based questions in this test are are 2, 4, 7, 9, 15, 19, 20, 23, 25, 28, 30, 36, 40, and 41. Finally the stress based questions are 1, 6, 8, 11, 12, 14, 18, 22, 27, 29, 32, 33, 35 and 39.

Each of the questions will be summed per category to come up with the total score per category per respondent.

Once we have derived this information, we will drop all of the Q#A, Q#I, and Q#E columns, as well as the time related columns, since they will not contain anymore information of interest to this analysis.

```python
# list of column names of questions related to depression
depression_answers = ['Q3A', 'Q5A', 'Q10A', 'Q13A', 'Q16A', 'Q17A', 'Q21A', 'Q24A', 'Q26A', 'Q31A', 'Q34A', 'Q37A', 'Q38A', 'Q42A']

# list of column names of questions related to anxiety
anxiety_answers = ['Q2A', 'Q4A', 'Q7A', 'Q9A', 'Q15A', 'Q19A', 'Q20A', 'Q23A', 'Q25A', 'Q28A','Q30A', 'Q36A', 'Q40A', 'Q41A']

# list of column names of questions related to stress
stress_answers = ['Q1A', 'Q6A', 'Q8A', 'Q11A', 'Q12A', 'Q14A', 'Q18A', 'Q22A', 'Q27A', 'Q29A', 'Q32A', 'Q33A', 'Q35A', 'Q39A']

# create lists of time values for depression, anxiety and stress by taking the already initiated answer lists and getting their Q#E value instead
depression_time = [i.strip('A') + 'E' for i in depression_answers]
anxiety_time = [i.strip('A') + 'E' for i in anxiety_answers]
stress_time = [i.strip('A') + 'E' for i in stress_answers]
```

```python
# creating new columns of depression, anxiety and stress sums per respondent
data['dep_score'] = data[depression_answers].sum(axis=1)
data['anx_score'] = data[anxiety_answers].sum(axis=1)
data['str_score'] = data[stress_answers].sum(axis=1)

# create a list of all question columns
all_questions = depression_answers + anxiety_answers + stress_answers

# get the list of all 'question position' variables which follow the Q#I format
question_position = [i.strip('A') + 'I' for i in all_questions]

# drop all question variables which are now uneccessary
data = data.drop(columns=all_questions + depression_time + anxiety_time + stress_time + question_position
                         + ['introelapse', 'testelapse', 'surveyelapse'], errors = 'ignore')
```

### Depression, Anxiety and Stress Categories

Here we will create a column to classify the level of depression, anxiety and stress of the respondent per the [scoring guide](https://neurocogsystem.com/wp-content/uploads/2021/02/DASS-42-Scoring.pdf).

For _depression_, the score classifications are as follows:
   * **Under 10:** Not depressed
   * **10-13:** Mildly depressed
   * **14-20:** Moderately depressed
   * **21-27:** Severely depressed
   * **28+:** Extremely severe depression
    
For _anxiety_ the classifications are:
   * **Under 8:** Not anxious
   * **8-9:** Mildly anxious
   * **10-14:** Moderately anxious
   * **15-19:** Severely anxious
   * **20+:** Extremely severe anxiety
    
Finally, for _stress_ the classifications are:
   * **Under 15**: Not stressed
   * **15-18**: Mildly stressed
   * **19-25**: Moderately stressed
   * **26-33:** Severely stressed
   * **34+:** Extremely severe stress

```python
data['dep_cat']=np.select([data['dep_score'] < 10, 
                           data['dep_score'].between(10,14),
                           data['dep_score'].between(14,21),
                           data['dep_score'].between(21,28),
                           data['dep_score'] >= 28],
                           ['None', 'Mild', 'Moderate', 'Severe', 'Extreme'])

data['anx_cat'] = np.select([data['anx_score'] < 8, 
                             data['anx_score'].between(8,10),
                             data['anx_score'].between(10,15),
                             data['anx_score'].between(15,20),
                             data['anx_score'] >= 20],
                             ['None', 'Mild', 'Moderate', 'Severe', 'Extreme'])

data['str_cat'] = np.select([data['str_score'] < 15, 
                             data['str_score'].between(15,19),
                             data['str_score'].between(19,26),
                             data['str_score'].between(26,34),
                             data['str_score'] >= 34],
                             ['None', 'Mild', 'Moderate', 'Severe', 'Extreme'])
```

### TIPI Classifications

The methodology for the TIPI or Ten-Item Personality Inventory can be found [here](http://gosling.psy.utexas.edu/wp-content/uploads/2014/09/tipi.pdf)

The categories assessed in this test are:
   * Extraversion
   * Agreeableness
   * Conscientiousness
   * Emotional Stability
   * Openness to Experiences

There are two questions per category. One where the item is assessed by the number the respondent put, and one that is it's opposite where the number is inversed. Since these are 7 levels in this survey, the inverse would be the number the survey respondent entered subtracted from 7. The questions for each category are:
   * _Extraversion:_ 1 and 6 reversed
   * _Agreeableness:_ 7 and 2 reversed
   * _Conscientiousness:_ 3 and 8 reversed
   * _Emotional Stability:_ 9 and 4 reversed
   * _Openness to Experiences:_ 5 and 10 reversed

In order to assess the traits of the survey respondent one should take the face value of the first question and average it with the reversed value of the second question. We will create a column for each trait and input the average that we come to for each respondent. After this we will drop the individual TIPI questions in favor of the score per trait.

```python
# a list of the tipi questions that are taken at face value
normal_tipis = ['TIPI' + str(i) for i in range(1,11,2)]

# a list of the tipi questions that are reversed
reverse_tipis = ['TIPI' + str(i) for i in range(1,11) if i % 2 == 0]
```

```python
# reverse the value for the appropriate tipis
data[reverse_tipis] = 7 - data[reverse_tipis]
```

```python
# a list of each personality trait
personalities = ['extraversion', 'agreeableness', 'conscientiousness', 'stability', 'openness']

# get the average for the appropriate questions. Each are 5 digits away from eachother
for i in personalities:
    data[i] = (data['TIPI' + str(personalities.index(i) + 1)] + data['TIPI' + str(personalities.index(i) + 6)]) / 2 
```

```python
# drop the tipi questions now that we have the other scores
data = data.drop(columns=normal_tipis + reverse_tipis)
```
